```python
#Checking for Null Values:

df.isna().sum()
```

Observations: Only 'Review Rating' feature has null values, we shall fill them with median of their respective categories.

```python
#Filling null values

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform( lambda x: x.fillna(x.median()))
df[['Review Rating', 'Category']].head()
```

```python
#Checking again for null:

df.isna().sum()
```

All the column names need to be formatted into snake-casing for easy of coding and reference to other softwares (SQL for data analysis).

```python
df.columns
```

```python
df.columns = df.columns.str.lower()   
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#Column purchase_amount_(usd) still needs to be formatted:

df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

#Feature Engineering

```python
#Age group columns from 'age':

bins = [0, 25, 40, 55, 100]
labels = ["Young-Adult", "Adult", "Middle-Aged", "Senior"]

df["age_group"] = pd.cut(
    df["age"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

df[["age", "age_group"]].head()
```

```python
#Unique values in 'frequency_of_purchases':

df['frequency_of_purchases'].unique()
```

```python
#Purchase frequency days column from 'frequency_of_purchases'

mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
    
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(mapping)
df[['purchase_frequency_days', 'frequency_of_purchases']].head(3)
```

```python
#We investigate 'discount_applied' and 'promo_code_used' since they have redundant head values.

df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#Observation: Both features have exactly same values, hence we drop one feature:

df.drop('promo_code_used', axis = 1, inplace = True)
df.columns
```
